## MadaLine:

```python
data22 = pd.read_csv('MadaLine.csv')
d0 = data22[data22.iloc[:, 2] == 0]
x0 = d0.iloc[:, 0].values
y0 = d0.iloc[:, 1].values
d0 = list(zip(x0, y0))
d1 = data22[data22.iloc[:, 2] == 1]
x1 = d1.iloc[:, 0].values
y1 = d1.iloc[:, 1].values
d1 = list(zip(x1, y1))
pp1=plt.scatter(x0, y0, c="red")
pp2=plt.scatter(x1, y1, c="blue")
xd = np.array([-2, 2])
plt.title(f'MadaLine')
plt.xlabel("x")
plt.ylabel("y")
plt.legend((pp1, pp2),('Class 1', 'Class 2'))
plt.show()
```

```python
class MadalineModel:
    def __init__(self,numberneurons):
        self.numberneurons=numberneurons
        self.learningrate=learningrate=0.01
        self.weight=np.random.random([2,numberneurons])
        self.weightout=np.array([1/numberneurons]*numberneurons)
        self.bias=np.random.random(numberneurons)
        self.biasout=(numberneurons-1)/numberneurons
        self.error=[]
    def Madalinetrain(self,data,targte,epoch):
        data,targte=shuffle(data,targte)
        for e in range(epoch):
            target=[]
            net=[]
            for d, trgt in zip(data, targte):
                on,outy=self.calculateneuron(d)
                if trgt !=outy[0]:
                  if trgt==1:
                      reng=-1
                      max=-2
                      for i in range(len(on)):
                          if max<on[i]:
                              max=on[i]
                              reng= i
                      self.weight[0,reng]+=self.learningrate*(1-on[reng])*d[0]
                      self.weight[1,reng]+=self.learningrate*(1-on[reng])*d[1]
                      self.bias[reng]+=self.learningrate* (1- on[reng])
                  elif trgt== -1:
                      for nn in range(self.numberneurons):
                          if on[nn]>0:
                              self.weight[0,nn]+=self.learningrate*(-1-on[nn])*d[0]
                              self.weight[1,nn]+=self.learningrate*(-1-on[nn])*d[1]
                              self.bias[nn]+=self.learningrate*(-1 -on[nn])
                target.append(trgt)
                net.append(outy[0])
            self.error.append(self.cost(target,net))

            if self.error[-1]<0.1 :
              print(f'epoch = {e}')
              break
        #print(self.error)
        acc=[]
        for item in self.error:
          acc.append((1-item)*100)
        return acc #acc[-1]
    def activatefunc(self,net):
        if net>=0:
            return 1
        else:
            return -1
    def calculateneuron(self,data):
        cn=(data@self.weight)+self.bias
        nout=[self.activatefunc(n) for n in cn]
        oy=(nout@self.weightout)+self.biasout
        outy=self.activatefunc(oy)
        return np.array(cn),np.array([outy])
    def cost(self,target,net):
        return np.mean((1/2)*(np.array(target)-np.array(net))**2)

def line(x,wx,wy,i):
    return (-wx/wy)*x-i/wy
```

```python
# nouron=3:
d=[-1 for i in range(len(d0))]
d.extend(1 for i in range(len(d1)))
data=[]
data.extend(d0)
data.extend(d1)

A= True
c=0
r=[]
m=0
while A==True:
  model_3n=MadalineModel(numberneurons=3)
  r=model_3n.Madalinetrain(data,d,epoch=1000)
  c+=1
  if r[-1]>90 or c>99:
    A=False

if r[-1]>90:
  print(f'after {c} ran reach upon %90 accuracy')
  plt.plot(range(len(model_3n.error)),r)
  plt.title(f'Accuracy')
  plt.xlabel("epoch")
  plt.ylabel("accuracy")
  plt.show()
  print(f'Accuracy= %{r[-1]:.3f}')
else :
  print('epoch = 1000')
  print(f'after {c} ran reach   %{r[-1]:.3f} accuracy')
  plt.plot(range(len(model_3n.error)),r)
  plt.title(f'Accuracy')
  plt.xlabel("epoch")
  plt.ylabel("accuracy")
  plt.show()
  max_index = r.index(max(r))
  print(f'Max Accuracy= %{max(r):.3f} occur in epoch= {max_index}')

x = np.linspace(-1.5, 1.5, 100)
xd = np.array([-2, 2])

for i in range(model_3n.numberneurons):
    plt.plot(x, line(x, model_3n.weight[0][i],model_3n.weight[1][i],model_3n.bias[i]), "k", lw=1, ls="--",alpha=0.7)
    plt.xlim(-2, 2)
    plt.ylim(-2, 2)
print("-----------------------------------------------------")
pp1=plt.scatter(x0,y0,color="red")
pp2=plt.scatter(x1,y1,color="blue")
plt.title(f'MadaLine - number of neurons= 3')
plt.xlabel("x")
plt.ylabel("y")
xd=np.array([-2,2])
plt.legend((pp1,pp2),('Class 1','Class 2'))
plt.show()
```
